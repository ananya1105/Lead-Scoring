```python
#Objective: Remove existing customers from the third party channels

import pandas as pd
import matplotlib.pyplot as plt
import re
#Metabase Data files
gen = pd.read_csv('general_data.csv')
spend = pd.read_csv('spends_data.csv')
trips = pd.read_csv('trips_data.csv')
countries = pd.read_csv('countries_data.csv')
#csv files
LMS=pd.read_csv('data/LMS.csv')#no from city data
Ixi =pd.read_csv('data/Ixigo.csv')#from airport form
Tri =pd.read_csv('data/Tripoto.csv')#from present
Hol =pd.read_csv('data/HolidayIQ.csv')
Tt = pd.read_csv('data/Travel_Triangle.csv')
IEd =pd.read_csv('data/India Education.csv')

Met = pd.read_csv('data/Mentorix.csv')
ph = pd.read_csv('ph.csv')
ph.rename(columns={'phone_number':'Phone'},inplace = True)
LMSm = pd.merge(LMS,ph,how='left',on='Phone')
Ixim = pd.merge(Ixi,ph,how ='left',on='Phone')
Trim = pd.merge(Tri,ph,how='left',on='Phone')
Holm = pd.merge(Hol, ph, how = 'left', on='Phone')
IEdm = pd.merge(IEd, ph, how = 'left', on = 'Phone')
Metm = pd.merge(Met,ph,how='left',on='Phone')

ph['Phone'] = ph['Phone'].astype(str)
Ttm = pd.merge(Tt, ph, how ='left', on='Phone')

LMSm.fillna(0,inplace = True)
Ixim.fillna(0,inplace = True)
IEdm.fillna(0,inplace = True)
Metm.fillna(0,inplace = True)
Trim.fillna(0,inplace = True)
Holm.fillna(0,inplace = True)
Ttm.fillna(0,inplace= True)

LMSm.drop(LMSm[LMSm['crn']!=0].index,inplace=True)
Ixim.drop(Ixim[Ixim['crn']!=0].index,inplace=True)
Trim.drop(Trim[Trim['crn']!=0].index,inplace=True)
Holm.drop(Holm[Holm['crn']!=0].index,inplace=True)
IEdm.drop(IEdm[IEdm['crn']!=0].index,inplace=True)
Metm.drop(Metm[Metm['crn']!=0].index,inplace=True)
Ttm.drop(Ttm[Ttm['crn']!=0].index,inplace=True)

LMSm.reset_index()
Ixim.reset_index()
Trim.reset_index()
Holm.reset_index()
IEdm.reset_index()
Metm.reset_index()


LMS = LMSm.drop(['crn','utm_source','age','acquisition_source','address_state','address_city','address_pincode'],axis = 1)
Hol = Holm.drop(['crn','utm_source','age','acquisition_source','address_state','address_city','address_pincode'],axis = 1)
Ixi = Ixim.drop(['crn','utm_source','age','acquisition_source','address_state','address_city','address_pincode'],axis = 1)
Met = Metm.drop(['crn','utm_source','age','acquisition_source','address_state','address_city','address_pincode'],axis = 1)
Tri = Trim.drop(['crn','utm_source','age','acquisition_source','address_state','address_city','address_pincode'],axis = 1)
Tt = Ttm.drop(['crn','utm_source','age','acquisition_source','address_state','address_city','address_pincode'],axis = 1)
IEd = IEdm.drop(['crn','utm_source','age','acquisition_source','address_state','address_city','address_pincode'],axis = 1)


LMS.to_csv('data/LMS.csv')
Hol.to_csv('data/Hol.csv')
Ixi.to_csv('data/Ixi.csv')
Met.to_csv('data/Met.csv')
Tri.to_csv('data/Tri.csv')
Tt.to_csv('data/Tt.csv')
IEd.to_csv('data/IEd.csv')
```
